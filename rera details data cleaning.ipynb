```python
pattern = r'.*\b[0-9]+bhk\b.*'
new_df = df[df['apartment_type'].astype(str).str.match(pattern, na=False)]
```

```python
pattern1 = r'(\d+)\s?bhk[A-Za-z0-9]*'
replacement1 = r'\1 BHK'
new_df['apartment_type'] = new_df['apartment_type'].str.replace(pattern1, replacement1, case = False, regex=True)
```

```python
new_df
```

```python
def regex(new_df):
    d = new_df['apartment_type']
    str = ""
    elements = re.findall(r'([a-zA-Z\. ])?([\.])?(([\d])+(([\.])?([\d])+)?(\s{0,}|[-]|\.)+(bhk|rk))+([\w,.&])?',d,re.X|re.M|re.I)
    lenght = len(elements[0][2])
    a = elements[0][2]
    result = re.sub("-", " ", a)
    if lenght in range(5,8):
        return result
    else:
        str = d
        return str
```

```python
new_df = new_df.apply(lambda row : regex(row),axis=1)
```
